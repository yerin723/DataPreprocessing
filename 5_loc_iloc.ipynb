### DataFrame 인덱싱

loc - 라벨기반 추출

```python
# df.loc[행 인덱스 라벨, 열 이름]
```

- 예를 들어 df.loc[3, 'math']와 같이 코드를 작성하면 행 인덱스가 3이고 열 이름이 'math'인 데이터를 추출할 수 있다.

iloc - 인덱스기반 추출

```python
# df.iloc[행 인덱스, 열 인덱스]
```

- 예를 들어 df.iloc[0, 1]과 같이 코드를 작성하면 첫 번째 행(0)에서 두 번째 열(1) 데이터를 추출할 수 있다.
- iloc는 loc와 달리 인덱스 번호를 이용하기 때문에  데이터프레임의 구조가 바뀌거나 열 이름이 변경되더라도 데이터 추출 방식이 유지된다.

```python
import pandas as pd
df = pd.DataFrame({'a':[i for i in range(1, 11)], 'b':[i for i in range(11,21)], 'c':[i for i in range(21,31)]})
df
```

```python
# a, b열 추출 (두 개 이상의 열을 추출하려면 열 이름을 리스트로 묶어야한다.)
df[['a', 'b']]
```

#### Ioc를 활용한 데이터 인덱싱 문제 1

```python
# 첫 번재 행의 데이터를 출력하시오.
df.loc[0,]
```

```python
# 2 ~ 4번째 행의 데이터를 출력하시오.
df.loc[2:4]
```

#### Ioc를 활용한 데이터 인덱싱 문제 2

```python
# g 인덱스의 데이터를 출력하고, c 인덱스부터 끝까지 출력하시오.
index = ['a', 'b', 'd', 'c', 'e', 'f', 'g', 'g', 'h', 'i']
df = pd.DataFrame({'num1':[i for i in range(1, 11)],
                   'num2':[i for i in range(11, 21)],
                   'num3':[i for i in range(21, 31)]},
                  index = index)
```

```python
df
```

```python
df.loc['g']
```

```python
df.loc['c':]
```

#### Ioc를 활용한 데이터 인덱싱 문제 3

```python
# 열이 a와 c이며 인덱스가 g와 i인 데이터를 추출하시오.
index = ['a', 'b', 'd', 'c', 'e', 'f', 'g', 'g', 'h', 'i']
df = pd.DataFrame({'a':[i for i in range(1, 11)],
                   'b':[i for i in range(11, 21)],
                   'c':[i for i in range(21, 31)]},
                  index = index)
```

```python
df
```

```python
df.loc[['g','i'],['a','c']]
```

#### iIoc를 활용한 데이터 인덱싱 문제

```python
# 처음부터 다섯 번째 까지의 데이터와 첫 번째 열과 세 번째 열의 데이터를 추출하시오.
df.iloc[:5, [0, 2]] # 첫 번째 대괄호는 추출할 행의 범위, 두 번째 대괄호는 추출할 열의 번호를 넣는다.
```

#### 추가 데이터 인덱싱 문제

```python
# a가 3이상이고, b가 16미만이거나 c가 30인 데이터를 출력하시오.
result = df[(df['a'] >= 3) & ((df['b'] < 16) | (df['c'] == 30))]
result
```

### slice - 이름 또는 인덱스를 이용한 (부분)데이터 선택

```python
import pandas as pd
import numpy as np
from numpy.random import randn

# 연습 데이터 생성
np.random.seed(1234) # sample data 통일을 위해 seed 고정
df = pd.DataFrame(randn(10,2), columns=['A','B']) # 변수 2개, 자료수 10개
```

```python
df
```

```python
df['A'] # df의 A열을 가리킴
```

```python
df[:5] # 콜론을 이용해 행의 범위로 가져오기
```

### query - 조건을 이용한 데이터 선택

```python
df.query('A > 0.5')
```

```python
df.query('A > B') # df[df.A > df.B]와 동일
```

```python
df.query('A > 0.2 & B < 0.8') # 복수 조건 가능
```
