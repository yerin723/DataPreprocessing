### 정렬

1. 인덱스기준
- 인덱스만 초기화하며 값은 그대로 유지된다.

```python
import pandas as pd
df = pd.DataFrame({'a':[1,3,5,7,9], 'b':[4,7,8,9,10], 'c':[1,1,2,3,4]})
```

```python
# 인덱스 오름차순 정렬
df.sort_index(inplace=True) # inplace=True는 기존 인덱스 삭제)
df
```

```python
# 인덱스 내림차순 정렬
df.sort_index(ascending=False, inplace=True) 
df
```

```python
# 인덱스 초기화
df.reset_index(drop=True)
df
```

```python
# 값 기준 정렬
df = pd.DataFrame({'a':[3,1,5,0,9], 'b':[4,7,8,9,10], 'c':[1,1,2,3,4]})
df.sort_values(by=['a'], inplace=True) # a열 기준으로 오름차순 정렬
df
```

```python
df.sort_values(by=['b'], ascending=False, inplace=True) # b열 기준으로 내림차순 정렬
df
```

```python
df.sort_values(by=['a','b'], ascending=False, inplace=True) # a, b열 기준으로 내림차순 정렬
df
```

```python
df.sort_values(by=['c'], inplace=True) # c열 기준 오름차순 정렬
df.sort_values(by=['a'], ascending=False, inplace=True) # a열 기준 내림차순 정렬

df.reset_index(drop=True, inplace=True) # index 초기화
df
```
